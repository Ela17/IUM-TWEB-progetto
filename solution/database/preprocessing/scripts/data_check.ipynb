```python
import pandas as pd
from pathlib import Path
import os 
```

```python
base_dir = Path().resolve().parent.parent.parent
data_clean_path = base_dir / 'data_clean'
```

```python
data = {}
for csv in data_clean_path.glob('*.csv'):
    name = csv.stem
    data[name] = pd.read_csv(csv)
```

```python
multi_roles = data['actors'].groupby(['id_movie', 'name'])['role'].nunique().reset_index()

# Tieni solo quelli con piÃ¹ di un ruolo
multi_roles = multi_roles[multi_roles['role'] > 1]

print(multi_roles)
```

```python
for name,df in data.items():
    
    print(df.info())
    print('\n\n')
```

```python
for name,df in data.items():
    print('+++++',name,'+++++')
    for col in df.columns:
        print(f"\n{col}:")
        print(f"  Non-null: {df[col].count()}")
        print(f"  Null: {df[col].isnull().sum()}")
        print(f"  Unique values: {df[col].nunique()}")
    print(f"\n")
```
